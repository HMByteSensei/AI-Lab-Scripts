```python
# g)
import numpy as np

def kvadratna(M):
    n = len(M)
    korijen = int(np.sqrt(n))
    if korijen * korijen == n:
        return M.reshape((korijen, korijen))
```

```python
# 4.
# a)
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1, 100, 100)
x = np.sin(n)
y = np.cos(n)

plt.plot(n, x, label='sin(n)')
plt.plot(n, y, label='cos(n)')
plt.xlabel('n')
plt.ylabel('Vrijednost')
plt.title('Sinus i kosinus funkcije')
plt.legend()
plt.grid()
plt.show()
```

```python
# b)
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1, 100, 100)
x = np.sin(n)
y = np.cos(n)

plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(n, x, label='sin(n)')
plt.xlabel('n')
plt.ylabel('Vrijednost')
plt.title('Sinus funkcija')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(n, y, label='cos(n)', color='r')
plt.xlabel('n')
plt.ylabel('Vrijednost')
plt.title('Kosinus funkcija')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
